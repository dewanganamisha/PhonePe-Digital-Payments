1.3: Check for missing values

```python
def missing_perc(df) :
  missing_values = df.isnull().sum()
  missing_percentage = (missing_values/len(df))*100
  missing_data = missing_percentage[missing_percentage > 0]

  if not missing_data.empty:
    highest_missing_column = missing_data.idxmax()
    highest_missing_percentage = missing_data.max()

    print(f"Column with highest missing values: {highest_missing_column} {highest_missing_percentage:.2f}%\n")
```

Task 3: Data Quality Checks
3.1: Ensure data consistency across state and district levels

```python
state_aggregates = State_txn_users.groupby("State").agg(State_Total_Transactions=("Transactions", "sum"),State_Total_Transaction_Amount=("Amount (INR)", "sum"),State_Total_Registered_Users=("Registered Users", "sum"))
print(state_aggregates)
```

```python

district_aggregates = District_Txn_users.groupby("State").agg(
    District_Total_Transactions=("Transactions", "sum"),
    District_Total_Transaction_Amount=("Amount (INR)", "sum"),
    District_Total_Registered_Users=("Registered Users", "sum")
)
comparison_df = pd.merge(district_aggregates, state_aggregates, on="State", how="inner")
# Compare district-level and state-level aggregates
comparison_df["Transactions_Match"] = comparison_df["District_Total_Transactions"] == comparison_df["State_Total_Transactions"]
comparison_df["Transaction_Amount_Match"] = comparison_df["District_Total_Transaction_Amount"] == comparison_df["State_Total_Transaction_Amount"]
comparison_df["Registered_Users_Match"] = comparison_df["District_Total_Registered_Users"] == comparison_df["State_Total_Registered_Users"]
comparison_df.head()
```

```python
discrepancies = comparison_df[
    (comparison_df["District_Total_Transactions"] != comparison_df["State_Total_Transactions"]) |
    (comparison_df["District_Total_Transaction_Amount"] != comparison_df["State_Total_Transaction_Amount"]) |
    (comparison_df["District_Total_Registered_Users"] != comparison_df["State_Total_Registered_Users"])
]
print(discrepancies)
```

Task 4: Data merging and advanced analysis  
4.1: Ratio of users to population by state

```python
state_population = District_Demographics.groupby("State").agg(
    Total_Population=("Population", "sum")
).reset_index()
registered_users=State_txn_users.groupby("State").agg(Total_Registered_Users=("Registered Users","sum")).reset_index()
state_user_population_ratio = pd.merge(registered_users, state_population, on="State", how="inner")
```

```python
state_user_population_ratio["User_Population_Ratio"] = (
    state_user_population_ratio["Total_Registered_Users"] / state_user_population_ratio["Total_Population"]
)
state_user_population_ratio.head()
```

```python
state_user_population_ratio = state_user_population_ratio.sort_values(by="User_Population_Ratio", ascending=False)

# Plot column chart
plt.figure(figsize=(12, 6))
plt.bar(state_user_population_ratio["State"], state_user_population_ratio["User_Population_Ratio"], color="royalblue")
plt.xticks(rotation=90, ha='right')
plt.xlabel("State")
plt.ylabel("User to Population Ratio")
plt.title("Ratio of Registered Users to Population by State")
plt.show()
```

4.2: Correlate population density with transaction volume

```python
merged_data = pd.merge(District_Txn_users, District_Demographics, on=["State", "District"], how="inner")
merged_data = merged_data[merged_data["Area (sq km)"] > 0]
merged_data = merged_data.dropna(subset=["Density", "Transactions"])
correlation = merged_data["Density"].corr(merged_data["Transactions"])
print(round(correlation,3))
```

```python

sns.heatmap(merged_data.corr(numeric_only = True),annot=True)
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_data, x="Density", y="Transactions", alpha=0.7)

# Add title and labels
plt.title("Correlation between Population Density and Transaction Volume",fontsize=14)
plt.xlabel("Population Density (people per sq. km)",fontsize=12)
plt.ylabel("Transaction Volume",fontsize=12)
plt.grid(True)

# Show plot
plt.show()
```
